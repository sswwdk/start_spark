# RDD 객체 생성

```python
spark = SparkSession.builder.appName('pyspark example1').getOrCreate()
```

```python
# RDD 생성
rdd = spark.sparkContext.parallelize([1,2,3,4,5])
rdd
```

```python
data1
```

```python
# rdd 객체를 출력하는 함수 (N개를 무조건 지정해줘야 된다)
rdd.take(5)
```

```python
# map 연산 : rdd 값으로 연산
squared_rdd = rdd.map(lambda x : x*x)
squared_rdd
```

```python
squared_rdd.take(3)
```

```python
squared_rdd.collect()
```
